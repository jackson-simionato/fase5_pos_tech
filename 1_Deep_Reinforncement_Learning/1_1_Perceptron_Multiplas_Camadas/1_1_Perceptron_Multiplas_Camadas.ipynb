# Tratando a variável target

Target está como string. É necessária converter para valores binários.

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

dados.Class = label_encoder.fit_transform(dados.Class)
```

```python
dados
```

# Seleção de colunas para o modelo

Como falado anteriormente, algumas colunas do dataframe possuem correlação muito forte entre si, evidenciando a existência de variáveis redundantes.

Neste caso serão descartadas as variáveis:
- Área_Convexa
- Equiv_Diâmetro
- Maior_Eixo_Comprimento

# Split dos dados

```python
X = dados.drop(columns=['Class', 'Área_Convexa','Maior_Eixo_Comprimento','Equiv_Diâmetro'])
y = dados['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

# Pré-processamento

## Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Fit do scaler
scaler = StandardScaler()
scaler.fit(X_train)
scaler.fit(X_test)
```

```python
# Aplicação do scaler
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Reshape da variável target

Estrutura de dados atual -> unidimensional, como uma coluna

Estrutura de dados para deep learning -> array multidimensional, como um tensor

```python
import numpy as np
```

```python
y_train = np.asarray(y_train).astype('float32').reshape((-1,1))
y_test = np.asarray(y_test).astype('float32').reshape((-1,1))
```

```python
y_train
```

# Arquitetura da rede neural

Bibliotecas: keras e tensorflow

```python
import tensorflow as tf
from tensorflow import keras
from keras import models, layers, metrics
from keras.optimizers import Adam
from keras.metrics import Precision
from tqdm.keras import TqdmCallback
```

```python
# Definição SEED
tf.random.set_seed(7)

# Definindo shape das entradas e saídas da rede + tamanho batch processamento
input_shape = X_train.shape[1]
output_shape = y_train.shape[1]
batch_size = 20

# Abrindo sequencia de neuronios
model = models.Sequential()

# Input Layer
# Entrada da rede
model.add(layers.Dense(
                    batch_size,
                    input_shape=(input_shape,),
                    activation='relu'
))

# Hidden Layer 1
# Camada oculta 1
model.add(layers.Dense(
                    12,
                    activation='relu'
))

# Hidden Layer 2
# Camada oculta 2
model.add(layers.Dense(
                    6,
                    activation='relu'
))

# Dropout Layer -> evitar que alguns neuronios tenham muita influencia no resultado, neuronios desativados aleatoriamente ao longo do processo
# Aplicando regularização
model.add(layers.Dropout(0.5))

# Output Layer
# Camada de saída
model.add(layers.Dense(
                    output_shape,
                    activation='sigmoid'
))

# Configuração do otimizador Adam
learning_rate = 0.001
otimizador = Adam(learning_rate=learning_rate)

# Compilar modelo
model.compile(loss='binary_crossentropy', optimizer=otimizador, metrics=['accuracy'])

# Summary
model.summary()
```

```python
# Configurando épocas de processamento
epoch = 100

hist = model.fit(X_train,
                 y_train,
                 epochs=epoch,
                 batch_size=batch_size,
                 shuffle=True,
                 validation_data=(X_test, y_test),
                 # verbose = barra de processamento
                 verbose=0,
                 callbacks=[TqdmCallback(verbose=0)])
```

```python
acc = "{:.2%}".format(hist.history['accuracy'][-1])
print(acc)
```

# Validação do modelo

```python
acc = hist.history['accuracy']
val_acc = hist.history['val_accuracy']

loss = hist.history['loss']
val_loss = hist.history['val_loss']

epochs_range = range(epoch)

# Plot acurácia
plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Acurácia - train set')
plt.plot(epochs_range, val_acc, label='Acurácia - test set')
plt.legend(loc='lower right')
plt.title('Acurácia - train e test set')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label = 'Erro - train set')
plt.plot(epochs_range, val_loss, label = 'Erro - test set')
plt.legend(loc='lower right')
plt.title('Erro - train e test set')

plt.show()
```

Os resultados são satisfatórios, há um bom equilíbrio entre o erro e acurácio do modelo com relação aos conjuntos de treino e teste, além de um certo patamar estabilizado destes valores a partir da época 20. Neste sentido, valeria também o testo de um novo modelo com número menor de épocas, o que o tornaria mais eficiente.

```python
from sklearn.metrics import classification_report
```

```python
y_pred = model.predict(X_test)
y_pred_class = [round(x[0]) for x in y_pred]
y_test_class = y_test

class_names = []

for i in y.unique():
    class_names.append(label_encoder.inverse_transform([i])[0])

print(classification_report(y_test_class, y_pred_class, target_names=class_names))
```
