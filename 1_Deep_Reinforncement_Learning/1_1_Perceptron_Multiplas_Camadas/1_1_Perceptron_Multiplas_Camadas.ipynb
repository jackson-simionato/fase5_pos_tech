# Análise exploratória

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Distribuição

```python
dados.hist(bins=50, figsize=(12,12))
plt.show()
```

Analisando os histogramas, é possível perceber que boa parte das variáveis possui distribuição próxima da normal, especialmente atributos como Área, Perímetro, Maior_Eixo_Comprimento, Proporção, etc.

Entretanto, algumas variáveis como Extensão, Excentricidade e principalmente Solidez possuem uma certa assimetria no gráfico de distribuição, indicando possível ocorrência de outliers.

## Correlação

```python
corr = dados.drop(columns=['Class']).corr().round(2)

fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(data = corr, annot=True, linewidths=.5, ax=ax)

plt.show()
```

Percebe-se alguns valores muito altos de correlação entre variáveis do dataframe, como é o caso da Área_Convexa e Equiv_Diâmetro, que possuem forte correlação com Área, Perímetro e atributos de comprimento do eixo das sementes. Tais variáveis estão descrevendo propriedades relacionadas aos tamanhos das sementes, portanto podem não haver necessidade de manter todas elas no modelo de classificação.

Nas próximas etapas da construção do modelo, será avaliada a possibilidade de retirar algumas destas variáveis do modelo de redes neurais.

Os valores de correlação negativa também ajudam a entender como outras variáveis da base se relacionam. Por exemplo, a variável Proporção possui correlação negativa muito forte com Redondeza e Compacidade, já que sementes compridas e estreitas serão menos compactas e arredondadas.
